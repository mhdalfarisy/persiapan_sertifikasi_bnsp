# Feature enginering

```python
import pandas as pd
import numpy as np

# Simulasi dataset kecil
df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female'],
    'ContractType': ['Month-to-month', 'One year', 'Two year'],
    'CustomerLevel': ['Bronze', 'Silver', 'Gold'],  # Ordinal
    'Tenure': [5, 24, 36],  # Rasio (diskrit)
    'MonthlyCharges': [70.35, 99.20, 60.10],  # Rasio (continuous)
    'Churn': ['Yes', 'No', 'No']
})

print("DATA AWAL:")
print(df)
```

## NOMINAL → One Hot Encoding

```python
df_nominal = pd.get_dummies(df, columns=['ContractType'], drop_first=True)

print("\nSETELAH ONE HOT (Nominal):")
print(df_nominal)
```

## ORDINAL → Mapping

```python
ordinal_mapping = {
    'Bronze': 1,
    'Silver': 2,
    'Gold': 3
}

df_nominal['CustomerLevel'] = df_nominal['CustomerLevel'].map(ordinal_mapping)

print("\nSETELAH ORDINAL ENCODING:")
print(df_nominal)
```

## BINARY → Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df_nominal['Gender'] = le.fit_transform(df_nominal['Gender'])
df_nominal['Churn'] = le.fit_transform(df_nominal['Churn'])

print("\nSETELAH LABEL ENCODING (Binary):")
print(df_nominal)
```

## RASIO / INTERVAL → Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_nominal[['MonthlyCharges', 'Tenure']] = scaler.fit_transform(
    df_nominal[['MonthlyCharges', 'Tenure']]
)

print("\nSETELAH SCALING (Numerical):")
print(df_nominal)
```

### FINAL DATA SETELAH FEATURE ENGINEERING

```python
print("\nDATA FINAL SIAP MODELING:")
print(df_nominal)
```

| Tipe Data | Skala | Perlakuan (Feature Engineering) | Hasil Setelah FE |
|------------|--------|----------------------------------|-------------------|
| Nominal | Categorical | One Hot Encoding | Jumlah kolom bertambah (dummy 0/1) |
| Ordinal | Categorical | Mapping / Ordinal Encoding | Tetap 1 kolom numerik dengan urutan |
| Binary | Categorical | Label Encoding | Berubah menjadi 0 dan 1 |
| Rasio | Numerical | Scaling (StandardScaler/MinMax) | Nilai terstandarisasi |
