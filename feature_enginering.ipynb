🔵 A. CATEGORICAL → Perlakuannya
🔹 1️⃣ Binary (Gender, Churn)

Contoh:

Male / Female

Yes / No

✅ Perlakuan:

Label Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Churn'] = le.fit_transform(df['Churn'])

Kenapa?
Karena hanya 2 kategori → aman diubah jadi 0 dan 1.

🔹 2️⃣ Nominal (ContractType, InternetService)

Contoh:

Month-to-month

One year

Two year

✅ Perlakuan:

One Hot Encoding

df = pd.get_dummies(df, columns=['ContractType', 'InternetService'], drop_first=True)

Kenapa?
Karena tidak punya urutan.
Kalau pakai Label Encoding bisa bikin bias.

🔹 3️⃣ Ordinal (Jika Ada)

Misalnya:
Bronze < Silver < Gold

✅ Perlakuan:

Manual mapping

mapping = {'Bronze':1, 'Silver':2, 'Gold':3}
df['CustomerLevel'] = df['CustomerLevel'].map(mapping)

Kenapa?
Karena memang ada ranking.

🟢 B. NUMERICAL → Perlakuannya
🔹 4️⃣ Diskrit (Tenure)

Contoh:
Tenure = jumlah bulan

✅ Perlakuan opsional:

Bisa langsung dipakai

Bisa dibuat grouping (feature baru)

Contoh feature engineering:

df['TenureGroup'] = pd.cut(df['Tenure'],
                           bins=[0,12,24,60],
                           labels=['New','Mid','Loyal'])

Ini bisa membantu model menangkap pola.

🔹 5️⃣ Continuous (MonthlyCharges)

Contoh:
350.75

✅ Perlakuan:

Scaling (untuk model linear)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['MonthlyCharges']] = scaler.fit_transform(df[['MonthlyCharges']])

Kenapa perlu scaling?

Karena:
Model seperti Logistic Regression sensitif terhadap skala.

🎯 3️⃣ Feature Engineering Tambahan (Advanced)
🔹 Interaction Feature

Misalnya:

Customer dengan kontrak Month-to-month + biaya tinggi → cenderung churn

Bisa buat fitur:

df['HighCharge'] = df['MonthlyCharges'] > df['MonthlyCharges'].median()
🔹 Feature Selection

Menghapus fitur yang tidak relevan:

df.corr()

Buang fitur dengan korelasi sangat rendah terhadap target.

🎯 4️⃣ Ringkasan Perlakuan Feature Engineering
Jenis Data	Perlakuan
Binary	Label Encoding
Nominal	One Hot Encoding
Ordinal	Mapping / Label
Diskrit	Bisa langsung / grouping
Continuous	Scaling
Semua	Cek missing value
Semua	Cek outlier
