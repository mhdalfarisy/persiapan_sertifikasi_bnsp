# RINGKASAN UJIAN KOMPETENSI DATA SCIENCE

------------------------------------------------------------------------

# I. Menentukan Objektif Bisnis

## 🎯 Yang Harus Kamu Kuasai

### ✅ 1. Merumuskan Problem Statement

-   Apa masalah bisnisnya?
-   Kenapa penting diselesaikan?
-   Dampak finansialnya apa?

### ✅ 2. Menentukan Business Objective

-   Apa tujuan proyek?
-   Apa yang ingin dicapai?

### ✅ 3. Menentukan Success Metric (KPI)

-   Bagaimana cara mengukur keberhasilan?

### 📌 Contoh:

**Problem:** Customer churn tinggi\
**Objective:** Memprediksi customer churn\
**KPI:** - Accuracy ≥ 85% - Recall ≥ 80% - Penurunan churn minimal 10%

------------------------------------------------------------------------

# II. Menentukan Tujuan Teknis

## 🎯 Yang Harus Kamu Kuasai

### ✅ 1. Memahami Task Data Science

  Kasus Bisnis     Task
  ---------------- ----------------
  Churn            Classification
  Prediksi Harga   Regression
  Segmentasi       Clustering

### 📌 Contoh Mapping:

-   Churn → Classification\
-   Prediksi harga rumah → Regression\
-   Segmentasi pelanggan → Clustering

------------------------------------------------------------------------

# III -- VI. Data Understanding & Preparation

## 🔹 WAJIB KUASAI PANDAS

-   read_csv()
-   info()
-   describe()
-   value_counts()
-   groupby()
-   isnull()
-   dropna()
-   fillna()
-   drop_duplicates()

------------------------------------------------------------------------

## 🔹 Visualisasi

Gunakan: - matplotlib - seaborn

Yang harus bisa: - countplot - boxplot - histogram - heatmap korelasi

------------------------------------------------------------------------

## 🔹 Data Cleaning

Wajib bisa jelaskan: - Handling missing value - Outlier detection (IQR /
Z-score) - Encoding categorical - Feature scaling

------------------------------------------------------------------------

# VII. Feature Engineering

Harus bisa: - Label Encoding - One Hot Encoding - StandardScaler -
Feature selection (korelasi) - Membuat fitur baru (contoh: tenure_group)

------------------------------------------------------------------------

# VIII -- X. Modeling

## 📌 Untuk Classification

Wajib kuasai: - Logistic Regression - Decision Tree - Random Forest -
Confusion Matrix - Accuracy - Precision - Recall - F1-score

------------------------------------------------------------------------

## 📌 Untuk Regression

Wajib kuasai: - Linear Regression - MAE - MSE - R2 Score

------------------------------------------------------------------------

# XI. Evaluasi & Review

Harus bisa: - Cross Validation - GridSearchCV - Interpretasi hasil
model - Rekomendasi bisnis berdasarkan model

------------------------------------------------------------------------

# 🎯 Yang Paling Sering Ditanya Asesor

1.  Kenapa memilih model tersebut?
2.  Kenapa perlu scaling?
3.  Kenapa recall lebih penting dari accuracy?
4.  Apa dampak bisnis dari model ini?
5.  Jika model jelek, apa yang kamu lakukan?

------------------------------------------------------------------------

# Kesimpulan

Untuk lulus ujian kompetensi, kamu harus mampu: - Menghubungkan problem
bisnis dengan solusi data science - Membersihkan dan memahami data -
Membangun model yang tepat - Mengevaluasi model dengan metrik yang
sesuai - Memberikan rekomendasi bisnis berdasarkan hasil analisis
