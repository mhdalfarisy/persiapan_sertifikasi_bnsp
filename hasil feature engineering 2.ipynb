🚀 1️⃣ DATA AWAL
import pandas as pd
import numpy as np

# Simulasi dataset kecil
df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female'],
    'ContractType': ['Month-to-month', 'One year', 'Two year'],
    'CustomerLevel': ['Bronze', 'Silver', 'Gold'],  # Ordinal
    'Tenure': [5, 24, 36],  # Rasio (diskrit)
    'MonthlyCharges': [70.35, 99.20, 60.10],  # Rasio (continuous)
    'Churn': ['Yes', 'No', 'No']
})

print("DATA AWAL:")
print(df)
🔵 2️⃣ NOMINAL → One Hot Encoding

Kolom:

ContractType (Nominal)

df_nominal = pd.get_dummies(df, columns=['ContractType'], drop_first=True)

print("\nSETELAH ONE HOT (Nominal):")
print(df_nominal)
✅ Hasilnya jadi seperti ini:
ContractType_One year	ContractType_Two year
0	0
1	0
0	1

Jumlah kolom bertambah ✔
Tidak ada ranking ✔

🟠 3️⃣ ORDINAL → Mapping

Kolom:
CustomerLevel
Bronze < Silver < Gold

ordinal_mapping = {
    'Bronze': 1,
    'Silver': 2,
    'Gold': 3
}

df_nominal['CustomerLevel'] = df_nominal['CustomerLevel'].map(ordinal_mapping)

print("\nSETELAH ORDINAL ENCODING:")
print(df_nominal)
✅ Hasil:
CustomerLevel
1
2
3

Tetap 1 kolom ✔
Model paham ranking ✔

🔵 4️⃣ BINARY → Label Encoding

Kolom:

Gender

Churn (Target)

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df_nominal['Gender'] = le.fit_transform(df_nominal['Gender'])
df_nominal['Churn'] = le.fit_transform(df_nominal['Churn'])

print("\nSETELAH LABEL ENCODING (Binary):")
print(df_nominal)

Hasil misalnya:

Male = 1
Female = 0
Yes = 1
No = 0

🟢 5️⃣ RASIO / INTERVAL → Scaling

Kolom:

MonthlyCharges (continuous)

Tenure (diskrit numerik)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_nominal[['MonthlyCharges', 'Tenure']] = scaler.fit_transform(
    df_nominal[['MonthlyCharges', 'Tenure']]
)

print("\nSETELAH SCALING (Numerical):")
print(df_nominal)
🎯 FINAL DATA SETELAH FEATURE ENGINEERING

Sekarang tampilkan semua:

print("\nDATA FINAL SIAP MODELING:")
print(df_nominal)

Skala	Perlakuan	Hasil
