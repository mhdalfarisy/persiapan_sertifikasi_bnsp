# 📚 FUNDAMENTAL TIPE DATA – DATA SCIENCE

## 🧠 1️⃣ Apa Itu Target?

**Target** adalah variabel yang ingin kita prediksi.

Contoh pada dataset Customer Churn:

Churn

Artinya:
- Yes → Customer berhenti
- No → Customer tetap

Karena tujuan bisnisnya adalah:
Memprediksi customer churn

Maka:
Churn = TARGET  
Kolom lain = FEATURE

---

## 🧠 2️⃣ Jenis Data Secara Umum

Secara umum data terbagi menjadi dua kelompok besar:

1. Categorical (Kategorikal)
2. Numerical (Numerik)

---

# 🔵 A. DATA CATEGORICAL (DATA KATEGORI)

Data yang berbentuk label / kelompok.

Terdiri dari:
- Binary
- Nominal
- Ordinal

---

## 🔹 1️⃣ Binary

Binary = hanya memiliki 2 nilai unik

Contoh:
- Gender → Male / Female
- Churn → Yes / No

Ciri-ciri:
- Hanya 2 kategori
- Tidak ada tingkatan

Contoh encoding:
Male → 1  
Female → 0  

Kenapa aman?
Karena hanya dua pilihan.

---

## 🔹 2️⃣ Nominal

Nominal = kategori lebih dari 2 dan tidak punya urutan

Contoh:
ContractType:
- Month-to-month
- One year
- Two year

Apakah Two year lebih tinggi dari One year?
Tidak.
Itu hanya jenis kontrak berbeda.

Contoh lain:
- Warna: Merah, Biru, Hijau
- Kota: Jakarta, Bandung, Surabaya
- InternetService: DSL, Fiber optic, No

Karena tidak punya urutan,
Gunakan One Hot Encoding.

---

## 🔹 3️⃣ Ordinal

Ordinal = kategori yang memiliki urutan / ranking

Contoh:
Low  
Medium  
High  

atau

SD  
SMP  
SMA  
S1  

Ada urutan:
S1 > SMA > SMP

Contoh lain:
Customer Level:
Bronze  
Silver  
Gold  

Karena ada urutan, bisa pakai:
- Label Encoding
- Manual mapping

---

# 🟢 B. DATA NUMERICAL (DATA ANGKA)

Data berbentuk angka dan bisa dihitung secara matematis.

Terbagi menjadi:
- Diskrit
- Berkelanjutan (Continuous)

---

## 🔹 4️⃣ Data Diskrit

Data berupa angka bulat dan hasil hitungan.

Contoh:
- Jumlah anak
- Jumlah transaksi
- Tenure (jumlah bulan berlangganan)

Ciri:
- Tidak ada nilai pecahan
- Biasanya hasil menghitung

Contoh:
Tenure = 12 bulan

---

## 🔹 5️⃣ Data Berkelanjutan (Continuous)

Data angka yang bisa memiliki nilai pecahan (desimal).

Contoh:
- MonthlyCharges
- Berat badan
- Tinggi badan
- Suhu

Ciri:
- Bisa desimal
- Bisa diukur

Contoh:
MonthlyCharges = 350.75

---

# 📊 Contoh Identifikasi Dataset Customer Churn

Misalnya hasil df.info():

Gender              object  
ContractType        object  
InternetService     object  
MonthlyCharges      float64  
Tenure              int64

Klasifikasi:

Gender → Categorical → Binary  
ContractType → Categorical → Nominal  
InternetService → Categorical → Nominal  
MonthlyCharges → Numerical → Continuous  
Tenure → Numerical → Diskrit  
Churn → Categorical → Binary (Target)

---

# 🎯 Kenapa Penting Memahami Ini?

Karena salah memahami tipe data bisa menyebabkan:
- Salah encoding
- Model bias
- Accuracy rendah
- Interpretasi keliru

Contoh kesalahan:

Jika ContractType (Nominal) di-Label Encoding:

Month-to-month → 0  
One year → 1  
Two year → 2  

Model bisa menganggap:
Two year > One year > Month-to-month

Padahal itu tidak selalu benar.

---

# 🔥 Ringkasan Cepat

Target → Variabel yang ingin diprediksi  
Binary → 2 kategori saja  
Nominal → Banyak kategori, tidak ada urutan  
Ordinal → Ada urutan / ranking  
Diskrit → Angka bulat hasil hitungan  
Continuous → Angka desimal hasil pengukuran  

---

# 🎤 Jawaban Siap Ujian

Apa perbedaan nominal dan ordinal?

Nominal adalah kategori tanpa urutan seperti warna atau jenis kontrak.  
Ordinal memiliki tingkatan seperti level pendidikan atau rating.

Kenapa tidak semua categorical pakai Label Encoding?

Karena Label Encoding memberi representasi numerik yang bisa dianggap sebagai urutan oleh model. Untuk kategori nominal tanpa ranking, lebih tepat menggunakan One Hot Encoding agar tidak menimbulkan bias numerik.

# 🚀 Cara Cepat Identifikasi Saat Ujian

2. Apakah ada urutan? → Ordinal  
3. Tidak ada urutan dan lebih dari 2? → Nominal  
4. Apakah angka hasil hitungan? → Diskrit  
5. Apakah angka bisa desimal? → Continuous
