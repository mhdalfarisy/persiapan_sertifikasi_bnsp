🧩 1️⃣ Data Cleaning (Fondasi)

Walaupun sering dipisah, ini bagian awal feature engineering.

✅ Handling Missing Value

Mean / Median (numerical)

Mode (categorical)

✅ Handling Outlier

IQR method

Z-score

Capping

🔵 2️⃣ Encoding (Untuk Categorical Data)

Digunakan agar model bisa membaca teks.

✔ Label Encoding

Untuk binary atau ordinal

✔ One Hot Encoding

Untuk nominal

✔ Ordinal Encoding

Untuk data berurutan

✔ Target Encoding (advanced)
🟢 3️⃣ Scaling (Untuk Numerical)

Digunakan terutama untuk model linear.

✔ StandardScaler

Mean = 0, Std = 1

✔ MinMaxScaler

0–1 range

✔ RobustScaler

Tahan outlier

🟡 4️⃣ Feature Transformation

Mengubah distribusi data.

✔ Log Transformation

Untuk data skewed

✔ Square root

Untuk distribusi miring ringan

✔ Power Transformation

Contoh:

df['MonthlyCharges_log'] = np.log(df['MonthlyCharges'])
🟠 5️⃣ Feature Creation (Membuat Fitur Baru)

Ini yang bikin model naik level 🔥

Contoh:

✔ Binning / Grouping
df['TenureGroup'] = pd.cut(df['Tenure'],
                           bins=[0,12,24,60],
                           labels=['New','Mid','Loyal'])
✔ Interaction Feature
df['Charge_per_Tenure'] = df['MonthlyCharges'] / df['Tenure']
✔ Date Feature Extraction

year

month

day

weekday

🔴 6️⃣ Feature Selection

Mengurangi fitur yang tidak penting.

✔ Correlation filter
✔ SelectKBest
✔ Recursive Feature Elimination (RFE)
✔ Feature Importance (Random Forest)
🟣 7️⃣ Handling Imbalance (Untuk Target)

Jika target tidak seimbang:

✔ Oversampling (SMOTE)
✔ Undersampling
✔ Class Weight
⚫ 8️⃣ Dimensionality Reduction (Advanced)

Jika fitur terlalu banyak.

✔ PCA
✔ LDA
